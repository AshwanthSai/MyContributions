```python
from plotly.offline import init_notebook_mode, iplot
```

```python
import plotly.graph_objs as go
```

```python
import numpy as np
import pandas as pd

first_csv = pd.read_csv(r"C:\Users\tonyj\Desktop\Gun Data Set\gun-violence-data_01-2013_03-2018.csv");

my_sheet = "incidents";
file_name = r"C:\Users\tonyj\Desktop\Gun Data Set\historical-enriched.xlsx"
second_csv = pd.read_excel(file_name, sheet_name = 1)
```

```python
second_csv.head()
```

```python
second_csv = second_csv.drop(['city_or_county','latitude','longitude','postcode'], axis=1)
```

```python
second_csv.head()
```

```python
first_csv = first_csv.drop(['incident_id','city_or_county','address','incident_url','source_url','incident_url_fields_missing','participant_age','participant_gender','participant_name','participant_relationship','participant_status','sources','state_house_district','state_senate_district'], axis=1)
```

```python
first_csv.head()
```

```python
first_csv = first_csv.drop(['congressional_district','gun_stolen','gun_type','incident_characteristics','latitude','location_description','longitude','n_guns_involved','notes','participant_age_group','participant_type'], axis=1)
```

```python
first_csv.head()
```

```python
second_csv = second_csv[['date','state','n_killed', 'n_injured']]
```

```python
second_csv.head()
```

```python
df_merged = first_csv.append(second_csv, ignore_index=True)
```

```python
df_merged.set_index('date')
```

```python
df_merged["date"]=df_merged["date"].astype('datetime64')
```

```python
df_merged.dtypes
```

```python
df_merged.set_index('date')
```

```python
df_merged.reset_index(inplace=True)
```

```python
df_merged['year'] = df_merged['date'].dt.year
```

```python
df_merged.head()
```

```python
df_merged.set_index('date')
```

```python
df = df_merged;
```

```python
# Ommit 2013, lot of missing values
temp = df[df['year'] == 2013].groupby('date').agg({'state' : 'count', 'n_killed' : 'sum', 'n_injured' : 'sum'}).reset_index().rename(columns={'state' : 'incidents'})

trace1 = go.Scatter(x = temp.date, y = temp.incidents, name='Total Incidents', mode = "lines", marker = dict(color = '#cccccc'))
trace2 = go.Scatter(x = temp.date, y = temp.n_killed, name="Total Killed", mode = "lines", marker = dict(color = '#ffe600'))
trace3 = go.Scatter(x = temp.date, y = temp.n_injured, name="Total Injured", mode = "lines", marker = dict(color = '#333333'))

data = [trace1, trace2, trace3]
layout = dict(height=350, title = 'Gun Violence Incidents - 2013', legend=dict(orientation="h", x=-.01, y=1), xaxis= dict(title='Date Time', ticklen= 1))
fig = dict(data = data, layout = layout)
iplot(fig)
```
